```python
class aRow():
    def __init__(self, tupleData):
        self.date = tupleData[0]
        self.station = tupleData[1]
        self.item = tupleData[2]
        self.hrs = []
        for i in range(24):
            self.hrs.append(tupleData[3+i])
class error():
    def __init__(self, date, station, item, hr):
        self.date = date
        self.station = station
        self.item = item
        self.hr = hr
```

```python
import MySQLdb

db=MySQLdb.connect(host="localhost",user="root", passwd="swater0", db="airdb", charset="utf8")
cursor = db.cursor()
queryString = "SELECT * from air "
data = cursor.execute(queryString)
results = cursor.fetchall()

days = []
for row in results:
    aDay = aRow(row)
    days.append(aDay)
db.close()
print('fetch finish')
```

```python
import math
errors = []
errorMachine = []
errorProgram = []
errorManual = []
blank = []
error999 = []

for i in range(len(days)):
    data = days[i]
    for idx in range(24):
        try:
            x=float(data.hrs[idx])
            if math.isnan(x)==True:
                anError = error(data.date, data.station, data.item, str(idx))
                errors.append(anError)
            if '999' in data.hrs[idx]:
                anError = error(data.date, data.station, data.item, str(idx))
                error999.append(anError)
            if 'NR' not in data.hrs[idx]:
                float(data.hrs[idx])
                        
        except ValueError:
            anError = error(data.date, data.station, data.item, str(idx))
            errors.append(anError)
            
            val = data.hrs[idx]
            if '#' in val:
                errorMachine.append(anError)
            elif '*' in val:
                errorProgram.append(anError)
            elif 'x' in val:
                errorManual.append(anError)
            elif not val:
                blank.append(anError)
print('finished')


#\# 表示儀器檢核為無效值，
#\* 表示程式檢核為無效值，
#\x 表示人工檢核為無效值，
#\NR 表示無降雨，
#\空白 表示缺值。
#\風向資料888代表無風，
#\999則代表儀器故障。
```

```python
###save error data###
db=MySQLdb.connect(host="localhost",user="root", passwd="swater0", db="airdb", charset="utf8")
cursor = db.cursor()

errWriteCNT = 0
for idx in range(len(errors)):
    row = errors[idx]
    qString = "INSERT INTO errordata VALUES ('"+row.date+"','"+row.station+"','"+row.item+"','"+row.hr+"')"
    cursor.execute(qString)
    errWriteCNT += 1
    print(str(errWriteCNT/len(errors)))
db.close()
```

```python
import datetime
db=MySQLdb.connect(host="localhost",user="root", passwd="swater0", db="airdb", charset="utf8")
cursor = db.cursor()
qString = "SELECT * FROM air"
cursor.execute(qString)
origindata = cursor.fetchall()
writeCNT = 0
prev_process = 0.0
for ori in origindata:
    result = cursor.execute("SELECT * FROM errordata WHERE date='"+ori[0]+"'"+\
                            "and station='"+ori[1]+"'"+\
                           "and item='"+ori[2]+"'")
    
    if result == 0:
        hrs = ''
        for i in range(3,27):
            hrs += ",'"+ori[i]+"'"
        qString = "INSERT INTO cleandata VALUES('"+ori[0]+"','"+ori[1]+"','"+ori[2]+"'"+hrs+")"
        #print(qString)
        cursor.execute(qString)
    
    writeCNT += 1
    process = round(writeCNT/len(origindata),2)
    if process != prev_process:
        print(str(process)[2:4]+"%     "+str(datetime.datetime.now()))
    prev_process = process
    




db.close()
print("filt finish")
```
